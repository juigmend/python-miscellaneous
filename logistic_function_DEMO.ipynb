```python
import numpy as np
import matplotlib.pyplot as plt

def log_func(x):
    return 1 / (1+np.e**(-x))

def logistic_function_demo(input_range, centre_offset, sigmoid_width, place_point):

    half_s_range = ( sigmoid_width/2 )
    sigmoid_minmax = [ -half_s_range , half_s_range ]
    alpha_offset = -( centre_offset * sigmoid_width / (input_range-1) )
    x_vector = np.linspace(sigmoid_minmax[0],sigmoid_minmax[1],input_range) + alpha_offset
    
    sigmoid = log_func(x_vector)
    
    half_input_range = int(input_range/2 - 1)
    effective_offset = np.where( np.max(np.diff(sigmoid)) == np.diff(sigmoid) )[0][0] - half_input_range 
    
    d_sigmoid = np.diff(sigmoid)
    d_sigmoid = d_sigmoid / np.max(d_sigmoid)
    
    print('offset [intended,effective] =',centre_offset,',',effective_offset)
    plt.plot(range(input_range),sigmoid, linewidth = 1.5 )
    plt.plot( d_sigmoid, ':c', linewidth = 1)
    if centre_offset == effective_offset:
        eff_mk_colour = '+b'
        int_mk_colour = '+b'
    else:
        int_mk_colour = 'Xy'
        eff_mk_colour = 'Xr'
    plt.plot( centre_offset+half_input_range , 0.5, int_mk_colour , markersize=16 )
    plt.plot( effective_offset+half_input_range , 0.5, eff_mk_colour , markersize=16 )

    point_x = ((place_point / input_range ) * sigmoid_width ) - sigmoid_minmax[1] + alpha_offset 
    point_y = log_func( point_x )
    plt.plot(place_point,point_y,'D', color=[0,0.8,0.5], markersize=10)

    plt.legend(['logistic function','derivative','intended centre','effective centre','placed point']) 
```

```python

input_range = 128 # <---
centre_offset = 0 # <---
sigmoid_width = 8 # <--- (inf...2), at 6 edges just become almost flat
place_point = 73  # <---

logistic_function_demo(input_range, centre_offset, sigmoid_width, place_point)
```
