```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel_length_x = 50 # <---
kernel_length_y = kernel_length_x * 2 # <---
sigma = kernel_length_x/2 # <---

kernel_length_x = int(kernel_length_x)
kernel_length_y = int(kernel_length_y)

xx = np.linspace(-kernel_length_x,kernel_length_x,kernel_length_x)
yy = np.linspace(-kernel_length_x,kernel_length_x,kernel_length_y)
xx, yy = np.meshgrid(xx, yy)
gauss_2D = 1. / (2. * np.pi * sigma**2) * np.exp(-(xx**2. / (2. * sigma**2.) + yy**2. / (2. * sigma**2.)))

plt.imshow(gauss_2D,aspect=1);
```
